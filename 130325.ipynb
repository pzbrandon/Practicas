```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x)**3
```

```python
X = np.linspace(-100,100,100)
```

```python
plt.plot(X,f(X))
```

```python
sp.integrate.quadrature(f,20,80)
```

```python
sp.integrate.trapezoid(f(X),X)
```

```python
for i in range(1,10):
    print(sp.special.roots_legendre(i))
```
