#### **1.** Escribe un programa que calcule una aproximación del número $\phi$, a partir de la siguiente serie

$$
    \phi = 2 \sum _{n=0}^{\infty} \frac{(-1)^n\pi^{2n}}{25^n(2n)!}
$$


```python
# Aquí irán las bibliotecas a usar
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defino dos funciones: la función factorial y la función que me de cada término de la suma y que depende de n

def factorial(x):
    if x==0 or x==1: 
        return 1
    elif x<0:
        return print('Error, número negativo')
    else:
        return x*factorial(x-1)

phi = 0
for n in range(0,50):
    phi = phi + ( ( ((-1)**n)*(np.pi)**(2*n) )/( (25**n)*factorial(2*n) ) )


print('El número aureo obtenido mediante la fórmula anterior a 50 términos es ',2*phi)
```

#### **2.** Usando la función _random.rand()_ de la biblioteca numpy, escribe un programa que simule una caminata aleatoria de 1000 pasos en una dimensión (derecha o izquierda) para $N$ pasos de longitud $l=1$ y que escriba en un archivo de texto el número de paso en una columna y la posición del caminante en otra.

```python
# Defino un arreglo que representa el número de pasos, desde el 1 hasta el 1000
n_paso = np.arange(1,1001,1)

# Como la función "random.rand" me devuelve números aleatorios entre el 0 y 1, defino una función que asigne lo siguiente
def binatoria(x):
    if (0<= x < 0.5 ): # A los números entre 0 y 0.5 les asigno el -1
        return -1
    elif (0.5 <= x <= 1): # A los números entre 0.5 y 1 les asigno el 1
        return 1
    else: print('idk') # Los demás casos no me intersan

# Con la función map puedo usar mi función anterior para un arreglo de 1000 términos aleatorios
pasos = list(map(binatoria,np.random.rand(1000)))
```

```python
# Guardo en un arreglo los pasos que voy a usar
P = np.array(pasos)

# Si el cuerpo comienza en una posición x = 0, su posición en el paso n va a ser la suma de esos n pasos.
# Así que para conocer la posición del cuerpo uso un for que sume los pasos y los vaya guardando en un arreglo

x = 0
posiciones = []
for i in P:
    x = x+i # A la posición anterior le sumo el paso
    posiciones.append(x) # Cada valor de la posición la voy guardando

# No recuerdo como escribir en un archivo de texto
mov = open("caminata.txt", 'w')
mov.write(P)
```
